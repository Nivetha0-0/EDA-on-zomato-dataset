```python
df1=pd.read_excel("Country-Code.xlsx")
df1.head()
```

```python
final=pd.merge(df,df1,on='Country Code',how='left')     #coz df has largedataset than df1
final.head()
```

```python
final.dtypes
```

```python
final.Country.value_counts()
```

```python
c_names=final.Country.value_counts().index
c_names
```

```python
c_vals=final.Country.value_counts().values
c_vals
```

```python
###FINDING TOP 3 COUNTRIES
plt.pie(c_vals[:3],labels=c_names[:3],autopct='%1.1f%%')#,startangle=90)
plt.show()
```

OBSERVATIONS:
1. India contributes the highest in Zomato followed by United States and United Kingdom

```python
final.columns
```

Analyzing numerical values

```python
rating=final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
#wherever column is named as 0 it is changed to Rating count
rating
```

OBSERVATIONS:
1. Ratings from 4.5 to 4.9 indicate Excellent


simiarly for all ranges

```python

import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=rating)
```

```python

import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=rating,palette=['blue','red','orange','yellow','green','green'])
```

OBSERVATIONS:
1. There is high number of non-rated reviews
2. Maximum number of ratings range from 3.0 to 3.8

```python
sns.countplot(x='Rating color',data=rating,palette=['blue','red','orange','yellow','green','green'])
```
